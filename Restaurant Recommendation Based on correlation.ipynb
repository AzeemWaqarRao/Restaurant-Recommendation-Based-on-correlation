# Restaurant Recommendation Based on correlation


### Importing Libraries

```python
import numpy as np
import pandas as pd
```

### Importing Data 

```python
data = pd.read_csv('rating_final.csv')
cuisine = pd.read_csv('chefmozcuisine.csv')
giodata = pd.read_csv('geoplaces2.csv', encoding = 'mbcs')
```

```python
data
```

```python
cuisine
```

```python
giodata
```

Getting only the we need from giodata

```python
places = giodata[['placeID','name']]
```

```python
places
```

### Finding Ratings of each restaurant 

Grouping data by places and finding counts of ratings

```python
ratings = pd.DataFrame(data.groupby('placeID')['rating'].count())
```

```python
ratings.head()
```

Adding a new column of average rating of each place

```python
ratings['rating_avg'] = pd.DataFrame(data.groupby('placeID')['rating'].mean())
ratings.sort_values('rating',ascending=False,inplace=True)
ratings
```

Adding Cuisine and place name to the ratings data frame

```python
ratings = pd.merge(ratings, places , on='placeID')
```

```python
ratings = pd.merge(ratings,cuisine,on='placeID')
```

```python
ratings
```

### Cross Tab

We we create a matrix in which their will be ratings by each user to each restaurant

```python
places_crosstab = pd.pivot_table(data=data, values='rating',index='userID',columns='placeID')
```

```python
places_crosstab
```

### Finding Correlation

Tortas Locas Hipocampo is the most rated restaurant <br />
we will find its coorelation with other restaurants based on user ratings

```python
tortas_rating = places_crosstab[135085]
tortas_rating = tortas_rating[tortas_rating >= 0]
```

```python
tortas_rating
```

```python
similar_to_Tortas = places_crosstab.corrwith(tortas_rating)
```

```python
corr_Tortas = pd.DataFrame(similar_to_Tortas, columns=['PearsonR'])
corr_Tortas.head()
```

```python
corr_Tortas.dropna(inplace=True)
corr_Tortas.head()
```

merging ratings dataframe with correlation data

```python
corr_Tortas = pd.merge(corr_Tortas, ratings, on='placeID')
```

```python
corr_Tortas.dropna(inplace=True)
```

```python
corr_Tortas.sort_values('rating', inplace=True, ascending=False)
```

```python
corr_Tortas
```

```python
corr_Tortas[corr_Tortas['Rcuisine'] == 'Fast_Food']
```

Based on these results we can recommend "Restaurante El Reyecito" to a user for having same cuisine and high coorelation with "Tortas Locas Hipocampo"
